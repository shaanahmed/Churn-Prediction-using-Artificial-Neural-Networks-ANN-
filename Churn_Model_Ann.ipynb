```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df.head()
```

```python
#Seperating dependent and independent variables
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
x.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

To increase the accuracy of the model we can increase the input nodes, or hidden laYER OR by using another activation function

```python
model=Sequential()
```

```python
model.add(Dense(13,activation='sigmoid',input_dim=11)) # 13 is the number of nodes
model.add(BatchNormalization())
model.add(Dense(13,activation='sigmoid')) # here we are adding one more layer in my ANN model
model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))
```

```python
#using Relu if we want to increase accuracy
model.add(Dense(13,activation='relu',input_dim=11)) # 13 is the number of nodes
model.add(Dense(13,activation='relu')) # here we are adding one more layer in my ANN model
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
mymodel=model.fit(x_train,y_train,epochs=30,validation_split=0.2)
```

```python
mymodel.history
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_pred=y_pred.argmax(axis=-1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
